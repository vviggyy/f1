```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load in files manually for now
#VET - Sebastian Vettel
#VER - Max Verstappen
#ALO - Fernando Alonso 

ym_2016_gap = pd.read_csv("YasMarina_2016_gap.csv")
ym_2016_place = pd.read_csv("YasMarina_2016_place.csv")
ym_2016_rs = pd.read_csv("YasMarina_2016_rs.csv")
ym_2016_gap_ver = pd.read_csv("YasMarina_2016_gap_ver.csv")
ym_2016_place_ver = pd.read_csv("YasMarina_2016_place_ver.csv")

ym_2016 = [ym_2016_gap, ym_2016_place, ym_2016_rs]
ym_2016_ver = [ym_2016_gap_ver, ym_2016_place_ver, ym_2016_rs]

m_2016_gap = pd.read_csv("Melbourne_2016_gap.csv")
m_2016_place = pd.read_csv("Melbourne_2016_place.csv")
m_2016_rs = pd.read_csv("Melbourne_2016.csv")
m_2016_gap_alo = pd.read_csv("Melbourne_2016_gap_alo.csv")
m_2016_place_alo = pd.read_csv("Melbourne_2016_place_alo.csv")
m_2016_gap_ver = pd.read_csv("Melbourne_2016_gap_ver.csv")
m_2016_place_ver = pd.read_csv("Melbourne_2016_place_ver.csv")


m_2016 = [m_2016_gap, m_2016_place, m_2016_rs]
m_2016_alo = [m_2016_gap_alo, m_2016_place_alo, m_2016_rs]
m_2016_ver = [m_2016_gap_ver, m_2016_place_ver, m_2016_rs]


cata_2016_gap = pd.read_csv("Catalunya_2016_gap.csv")
cata_2016_place = pd.read_csv("Catalunya_2016_place.csv")
cata_2016_rs = pd.read_csv("Catalunya_2016_rs.csv")
cata_2016_gap_alo = pd.read_csv("Catalunya_2016_gap_alo.csv")
cata_2016_place_alo = pd.read_csv("Catalunya_2016_place_alo.csv")
cata_2016_gap_ver = pd.read_csv("Catalunya_2016_gap_ver.csv")
cata_2016_place_ver = pd.read_csv("Catalunya_2016_place_ver.csv")

cata_2016 = [cata_2016_gap, cata_2016_place, cata_2016_rs]
cata_2016_alo = [cata_2016_gap_alo, cata_2016_place_alo, cata_2016_rs]
cata_2016_ver = [cata_2016_gap_ver, cata_2016_place_ver, cata_2016_rs]
```

```python
for iyear, (gap, place) in enumerate(zip(ym_gap, ym_place)):
    
    i = 1
    year = iyear + 2014
    plt.figure()
    cate = np.arange(1, 21, 1) #col names
    driver = gap.iloc[i,0]
    
    dist1 = gap.iloc[i, 1:]
    print(f"stdev_gap_{driver}: {np.std(dist1)}")
    dist1_avg = np.mean(dist1)
    
    dist2 = place.iloc[i, 1:]
    print(f"stdev_place_{driver}: {np.std(dist2)}\n\n")
    dist2_avg = np.mean(dist2)

    plt.bar(cate, dist1, color = "blue", alpha = 0.2, label = "GAP OF")
    plt.bar(cate, dist2, color = "red", alpha = 0.2, label = "PLACE OF")
    
    plt.title(f"{driver} positions (1000 simulations) {year}")
    plt.xlabel("place")
    plt.ylabel("number of placements")
    plt.legend()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(5, 10))
labels = ["(VET) YasMarina 2016 Gap Objective", "(VET) YasMarina 2016 Place Objective", "(VET) YasMarina 2016 Real Strategy"]
for i in range(len(ym_2016)):
    
    df = ym_2016[i]

    dist = df[df.iloc[:, 0] == "VET"].iloc[0,1:]
    cate = np.arange(1, len(dist)+1, 1)
    
    axes[i].bar(cate, dist)
    axes[i].set_title(labels[i])
    axes[i].set_xlabel("Position")
    axes[i].set_ylabel("Frequency of Placement")
    axes[i].set_xticks(np.arange(1, len(dist)+1, 1))  #evenly spaced ticks

    plt.subplots_adjust(wspace=0.3, hspace=0.5)  # Adjust horizontal and vertical space

plt.figure()

dist_gap = ym_2016_gap[ym_2016_gap.iloc[:, 0] == "VET"].iloc[0,1:]
dist_place = ym_2016_place[ym_2016_place.iloc[:, 0] == "VET"].iloc[0,1:]
diff = dist_gap - dist_place
cate = np.arange(1, len(dist)+1, 1)
c = ["green" if d > 0 else "red" for d in diff]

plt.bar(cate, dist_gap-dist_place, color = c)
plt.title("(VET) Difference in YasMarina 2016 placement frequency (Gap-Place)")
plt.xlabel("Position")
plt.ylabel("Frequency")
plt.xticks(np.arange(1, len(dist)+1, 1))
print(-np.sum(diff * cate))
```
